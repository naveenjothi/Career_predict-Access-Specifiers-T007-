# Creating or adding csv siles and accessing those csv's

```python
import pandas as pd
from IPython.display import display
import csv,io

def check(data):
    a=input("User: ")
    u=a.lower()
    for i in range(0,len(data)):
        if(data['dept'][i].lower()==u or data['short'][i].lower()==u or data['sh'][i].lower()==u):
            r=data['short'][i].lower()
            call(data,r)
            break
    else:
        b=input("Enter full form of the course: ")
        data['dept'].append(b)
        s=input("Enter the short form: ")
        ss=input("Enter some another short form: ")
        if(len(s)>=len(ss)):
            data['sh'].append(s)
            data['short'].append(ss)
        else:
            data['short'].append(s)
            data['sh'].append(ss)
        r=data['short'][-1].lower()
        call(data,r)
        
    return


def call(data,r):
    df=pd.DataFrame(data,columns=['dept','short','sh'])
    df.to_csv(r+'.csv',index=False)
    ca=pd.read_csv(io.StringIO(df.to_csv(index=False)))
    display(ca)
    return

data={'dept':['Information Technology', 'Computer Science Engineering', 'Electrical and Communication Engineering'],
       'short':['IT','CSE','ECE'],
        'sh':['Info Tech','Comp Eng','Elec and Comm Eng']
     }
check(data)
    
```

# Creating of csv files and entering questions and answers in runtime

```python
import pandas as pd
from IPython.display import display
import csv,io
from nltk.tokenize import word_tokenize

def csvcreater(data,r,li):
    r=r.lower()
    r=r.replace(" ", "")
    df=pd.DataFrame(data,columns=li)
    df.to_csv(r+'.csv',index=False)
    ca=pd.read_csv(io.StringIO(df.to_csv(index=False)))
    display(ca)
    return

def qnenter(data,r):
    li=['question','answer0']
    br=0
    while(1):
        qs=input("Question:")
        data['question'].append(qs)
        ans=input("Answer0:")
        data['answer0'].append(ans)
        a=int(input("If you want to continue answering press 1 end 0"))
        if(a==0):
            csvcreater(data,r,li)
            br=1
            break
        if(a==1):
            i=1
            while(1):
                data.update({'answer'+str(i):[]})
                answ=input("Answer"+str(i)+": ")
                data['answer'+str(i)].append(answ)
                li.append('answer'+str(i))
                i+=1
                b=int(input("If you want to continue answering press 1 for end 0"))
                if(b==1):
                    continue
                if(b==0):
                    csvcreater(data,r,li)
                    br=1
                    break
        if(br==1):
            break
data={'question':[],'answer0':[]}
r=input("Enter the file name: ")
qnenter(data,r)
```
